Na pasta "Empresas" vão os dados do portal de dados abertos, disponível [neste link](https://www.gov.br/receitafederal/pt-br/assuntos/orientacao-tributaria/cadastros/consultas/dados-publicos-cnpj)

```python
import pandas as pd
```

```python
pasta = 'empresas/'
complemento = '.EMPRECSV'
```

[Link para Dicionário dos dados](https://www.gov.br/receitafederal/pt-br/assuntos/orientacao-tributaria/cadastros/consultas/arquivos/NOVOLAYOUTDOSDADOSABERTOSDOCNPJ.pdf)

Natureza Jurídica "Organização Religiosa" :  3220

```python
lista = []

for i in range(1,10):
    path = pasta + str(i) + complemento
    df = pd.read_csv(path, sep=';', header = None)
    df.columns = ['CNPJ_basico', 'razao_social', 'natureza_juridica','qualificacao_do_responsavel', 'capital_social_da_empresa','porte_da_empresa','ente_federativo_responsavel']
    df = df.loc[df.natureza_juridica == 3220]
    lista.append(df)

dataframe = pd.concat(lista)
dataframe
```

```python
dataframe['CNPJ_basico'].value_counts()
```

```python
lista = []

for i in range(1,11):
    arq_path = pasta_estb +'csv/' + str(i) + '.csv'
    df = pd.read_csv(arq_path, sep=';', header = None, low_memory=False)
    lista.append(df)

dataframe_estb = pd.concat(lista)
dataframe_estb.head()
```

```python
lista = []

for i in range(1,10):
    path = pasta + str(i) + complemento
    df = pd.read_csv(path, sep=';', header = None)
    df.columns = ['CNPJ_basico', 'razao_social', 'natureza_juridica','qualificacao_do_responsavel', 'capital_social_da_empresa','porte_da_empresa','ente_federativo_responsavel']
#     df = df.loc[df.natureza_juridica == 3220]
    df = df.loc[df['CNPJ_basico'].isin(no_matches[0])]
    lista.append(df)

dfBusca = pd.concat(lista)
dfBusca
```
