```python
path = 'divida_ativa_uniao.csv'
df = pd.read_csv(path)
df
```

```python
df['valor'] = df['Todas Origens - Valor Consolidado da Inscrição'].apply(lambda x: x.replace('.','').replace(',','.')).astype(float)
```

```python
df['CNPJ_basico'] = df['CPF/CNPJ do Devedor Detalhado'].apply(lambda x: x.replace('.','')[0:8]).astype('int64')
```

```python
grupo_cnpj = df.groupby('CNPJ_basico')['valor'].sum().reset_index()
grupo_cnpj
```

```python
grupo_cnpj.info()
```

```python
def getDividas(cnpj):
    
    dividas = df.loc[df.CNPJ_basico == cnpj]
    
    nDividas = dividas.shape[0]
    nome = dividas.iloc[0]['Nome do Devedor Detalhado']       
    
    return [nome, nDividas, dividas.filter(['Ano da Inscrição', 'Número de Inscrição','Tipo da Situação da Inscrição', 'Tipo Inscrição',
       'Tipo Regularidade da Inscrição','valor']).to_dict('records')]
```

```python
# grupo_cnpj['Dividas'] = grupo_cnpj['CNPJ_basico'].apply(lambda x: getDividas(x))
```

```python
grupo_cnpj['Nome'] = grupo_cnpj['CNPJ_basico'].apply(lambda x: getDividas(x)[0])
grupo_cnpj['nDividas'] = grupo_cnpj['CNPJ_basico'].apply(lambda x: getDividas(x)[1])
grupo_cnpj['Dividas'] = grupo_cnpj['CNPJ_basico'].apply(lambda x: getDividas(x)[2])

grupo_cnpj['valor'] = round(grupo_cnpj['valor'],2)
```

```python
grupo_cnpj['nDividas'].sum()
```

```python
dez_mais = grupo_cnpj.sort_values(by='valor', ascending = False).head(10)
dez_mais.to_csv('dez_mais_divida_ativa.csv', index=False)
dez_mais
```

```python
def getNumDividas(ano):
    return df.loc[df['Ano da Inscrição'] == ano].shape[0]
```

```python
getNumDividas(2000)
```

```python
grupo_ano = df.groupby('Ano da Inscrição')['valor'].sum().reset_index()

grupo_ano['valor'] = round(grupo_ano['valor'],2)
grupo_ano['Número de Dívidas'] = grupo_ano['Ano da Inscrição'].apply(lambda x: getNumDividas(x))
grupo_ano['Somatório Valor'] = round(grupo_ano['valor'].cumsum(),2)
grupo_ano['Somatório Número de Dívidas'] = grupo_ano['Número de Dívidas'].cumsum()
grupo_ano.to_csv('divida_por_ano.csv', index=False)

grupo_ano
```
