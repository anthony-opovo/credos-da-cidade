## Buscando no DataSet Empresas linhas cuja razão social possui correspondênte ao dado da SEFIN:

```python
lista = []

for i in range(1,10):
    path = pasta + str(i) + complemento
    df = pd.read_csv(path, sep=';', header = None)
    df.columns = ['CNPJ_basico', 'razao_social', 'natureza_juridica','qualificacao_do_responsavel', 'capital_social_da_empresa','porte_da_empresa','ente_federativo_responsavel']
#     df = df.loc[df.natureza_juridica == 3220]
    df = df.loc[df['razao_social'].isin(dfSEFIN['razao_social'])]
    lista.append(df)

dfBusca = pd.concat(lista)
dfBusca
```

```python
lista_SEFIN_sem_CNPJ_correspondente =  dfSEFIN.loc[dfSEFIN.razao_social.isin(verifica.loc[verifica._merge == 'right_only']['razao_social'])]
lista_SEFIN_sem_CNPJ_correspondente.to_csv('lista_sefin_sem_CNPJ_correspondente_Empresas_Receita.csv')
lista_SEFIN_sem_CNPJ_correspondente
```

```python
dfSEFIN.loc[dfSEFIN.razao_social.isin(verifica.loc[verifica._merge == 'both']['razao_social'])]
```
